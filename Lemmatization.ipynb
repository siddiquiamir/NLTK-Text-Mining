```python
import nltk
```

```python
udhr= nltk.corpus.udhr.words('English-Latin1')
```

```python
print(udhr)
```

```python
udhr[:30]
```

```python
porter = nltk.PorterStemmer()
```

```python
[porter.stem(w) for w in udhr]
```

```python
wnl = nltk.WordNetLemmatizer()
```

```python
[wnl.lemmatize(w) for w in udhr]
```
